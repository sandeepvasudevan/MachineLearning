```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_diabetes
from sklearn.datasets import load_wine
```

```python
# Principal Component Analysis
dataset=load_wine()
data=pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
print('Shape of dataset',data.shape)
scaler=StandardScaler()
scaled_data=scaler.fit_transform(data)
principal=PCA(n_components=2)
X=principal.fit_transform(scaled_data)
print('Shape of dataset post PCA',X.shape)
print('Explained Varinace',principal.explained_variance_)
print('Explained Variance Ratio',principal.explained_variance_ratio_)
```

```python
#LinearDiscriminant Analysis
wine=load_wine()
dataset=pd.DataFrame(wine['data'],columns=wine['feature_names'])
target=wine['target']
print('Unique Targets',len(set(target)))
print('Dataset Shape',dataset.shape)
lda=LinearDiscriminantAnalysis()
X=lda.fit_transform(dataset,target)
print('Shape of LDA dataset',X.shape)


plt.xlabel('LDA1')
plt.ylabel('LDA2')
plt.scatter(X[:,0],X[:,1],c=target,cmap='rainbow')
```
